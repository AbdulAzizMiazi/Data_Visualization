```python
sns.set_theme(style="darkgrid")
```

```python
x = np.array([0, 2, 3, 4, 5, 6, 7, 8, 9 ,10, 60])
y = np.sin(x)

sns.lineplot(x=x, y=y)
plt.title("Line Plot")
plt.show()
```

#### Ploting with Data Sets

```python
# Load a default dataset
tips = sns.load_dataset('tips')
tips.head()
```

#### Line Plot

```python
sns.lineplot(x='total_bill', y='tip', data=tips)
plt.title("Line Plot Example")
plt.show()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="time")
plt.title('Scatter Plot with Color by Time')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="total_bill", y="tip", data=tips)
plt.title('Average Bill per Day')
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title('Boxplot of Total Bill per Day')
plt.show()
```

```python
flights = sns.load_dataset('flights')
# flights.head()
pivot_table = flights.pivot(index="month", columns="year", values="passengers")

sns.heatmap(pivot_table, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Heatmap of Passengers')
plt.show()
```
