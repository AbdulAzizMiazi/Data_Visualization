```python
import matplotlib.pyplot as plt
```

```python
# data
days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week
studying = [3, 4, 3, 5, 4, 3, 4]
playing = [2, 2, 1, 1, 2, 3, 2]
watching_tv = [2, 1, 2, 2, 1, 1, 1]
sleeping = [5, 5, 6, 5, 6, 5, 5]
```

```python
plt.figure(figsize=(8.5, 5.5))
plt.stackplot(days, 
            studying, playing, watching_tv, sleeping,
            labels=["Study", "Sports", "TV", "Sleep"],
            colors=["#F78D60", "#EA2264", "#640D5F", "#0D1164"],
            alpha=0.95)

plt.xticks(days, ["Mon", "Tue", "Wed", "Thur", "Fri", "Sat", "Sun"])

plt.xlabel("Day")
plt.ylabel("Activity (Hours)")
plt.legend(loc="upper right")
plt.grid(True)
plt.show()
```
