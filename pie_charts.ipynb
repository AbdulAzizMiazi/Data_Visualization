```python
import matplotlib.pyplot as plt
# plt.style.available
plt.style.use("ggplot")
# plt.rcParams['figure.figsize'] = (10, 6) 
```

```python
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
```

```python
plt.figure(figsize=(10, 8))
plt.pie(runs,
        labels=labels,
        # Labels & Percentage Placing
        autopct='%1.1f%%',
        pctdistance=1.2,
        labeldistance=1.1,
        # Esthetic Look of the Pie
        colors=["#320A6B", "#065084", "#0F828C", "#78B9B5"],
        explode=[0.08, 0, 0, 0],
        hatch=['**O', 'oO', 'O.O', '.||.'],
        # Highlighting the Edges of each slide
        wedgeprops={
            'edgecolor': '#D2C1B6',
            'linewidth': 1,
            'linestyle': '-'
            },
        # oriantation of the pie & 3D Effect
        startangle= -90,
        shadow=True
       )

plt.title("Career Runs of Indian Batsmen")
plt.legend()
plt.show()
```
