```python
import matplotlib.pyplot as plt
```

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 50, 55, 60, 65, 75, 85, 90]
```

```python
# Basic Scatter Plot
plt.scatter(study_hours, exam_scores)
plt.title("Students Study Time vs Exam Results")
plt.xlabel("Study Hours")
plt.ylabel("Exam Scrore")
plt.grid(True)
plt.show()
```

```python
# exam_scores
sizes = [score**1.3 for score in exam_scores]
colors = ["red" if score < 50 else "green" for score in exam_scores]

# plt.scatter(study_hours, exam_scores, s=20, c="blue")

plt.scatter(study_hours, exam_scores, s=sizes, c=colors)

plt.title("Colored & Sized Scatter Plot")
plt.xlabel("Study Hours")
plt.ylabel("Exam Scrore")
plt.grid(True)
plt.show()
```

```python
# Using a Colormap

plt.scatter(study_hours, exam_scores, s=sizes, c=exam_scores, cmap="plasma")
# here, "c" has to be a list numbers which is used by "cmap" to show the colors representing intensity

plt.colorbar()

plt.title("Scatter Plot with Colormap")
plt.xlabel("Study Hours")
plt.ylabel("Exam Scrore")
plt.grid(True)
plt.show()
```

```python
# Adding Annotations
plt.scatter(study_hours, exam_scores)

# Adding the labels (annotations)
for i in range(len(study_hours)):
    plt.annotate(f"S{i+1}", (study_hours[i]+0.09, exam_scores[i]-1.9))

plt.title("Scatter Plot with Annotations")
plt.xlabel("Study Hours")
plt.ylabel("Exam Scrore")
plt.grid(True)
plt.show()
```

```python
# Assume two groups: Class A and Class B
class_a_hours = [2, 4, 6, 8]
class_a_scores = [45, 55, 65, 85]

class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 50, 60, 70, 90]

plt.scatter(class_a_hours, class_a_scores, label='Class A', color='blue')
plt.scatter(class_b_hours, class_b_scores, label='Class B', color='orange')

plt.title('Scatter Plot: Class A vs Class B')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.legend()
plt.grid(True)
plt.show()
```
